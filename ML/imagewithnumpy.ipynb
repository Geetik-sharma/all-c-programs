```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1=np.array([1,2,3,4,5])
array2=np.random.rand(3,3)
array3=np.zeros((2,2))
```

saving the array to system

```python
np.save('array1.npy',array1)
np.save('array2.npy',array2)
np.save('array3.npy',array3)
```

reading the saved array from system

```python
loaded_array1=np.load('array1.npy')
print(loaded_array1)
```

```python
from PIL import Image
img=Image.open('download.png')
imgarray=np.array(img)
np.save('imgarray',imgarray)
```

```python
image=np.load('imgarray.npy')
print(image)
```

```python
try:
    image=np.load('imgarray.npy')
    #display
    plt.Figure(figsize=(10,10))
    plt.subplot(121)
    plt.imshow(image)
    plt.title('image')
    plt.grid(False)

    dark_img= 1-image
    plt.figure(figsize=(10,10))
    plt.subplot(121)
    plt.imshow(dark_img)
    plt.title("dark image")
    plt.grid(False)

except FileNotFoundError:
    print("file not found")
```
